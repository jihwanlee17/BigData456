```python
# KB Finance(105560)
df = fdr.DataReader('105560', '2020-01-01', '2020-12-31')
df.head(15)
```

```python
import pandas as pd

# KB Finance(105560), Hyundai Mtr(005380), SK Hynix(000660)
tiker_list = ['105560', '005380', '000660'] 
df_list = [fdr.DataReader(ticker, '2021-01-01', '2021-08-31')['Close'] for ticker in tiker_list]

len(df_list)
```

```python
# pd.concat()로 합치기

df = pd.concat(df_list, axis=1)
df.columns = ['KB Finance(105560)', 'Hyundai Mtr(005380)', 'SK Hynix(000660)']
df = df.dropna()
df.head(10)
```

```python
df.plot()
```

```python
df['KB Finance(105560)'].plot()
df['Hyundai Mtr(005380)'].plot(ls='--', color='r', secondary_y=True)
```

```python
# 정규화: 가격의 시작을 0 으로

df_norm = df / df.iloc[0] - 1
df_norm.plot()
```
