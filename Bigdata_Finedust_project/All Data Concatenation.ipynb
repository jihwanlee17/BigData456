```python
df0 = pd.DataFrame() #2001년부터 2013년까지 -> 동일한 형식, 분기별

for year in range(2001, 2014):
  for i in range(1,5):
    f=pd.read_excel(f'/content/drive/MyDrive/Colab Notebooks/Bigdata_finedust/{year}년0{i}분기.xlsx')
    dfa = pd.DataFrame(f) 
    df0 = df0.append(dfa, ignore_index=True)

```

```python
df1 = pd.DataFrame() #2014년부터 2016년까지 -> PM25변수만 새로 추가됨, 얘네끼리 형식 동일, 분기별

for year in range(2014, 2017):
  for i in range(1,5):
    if year != 2015:
      f=pd.read_csv(f'/content/drive/MyDrive/Colab Notebooks/Bigdata_finedust/{year}년 {i}분기.csv', encoding='cp949')
    else:
      f=pd.read_csv(f'/content/drive/MyDrive/Colab Notebooks/Bigdata_finedust/{year}년 {i}분기.csv')
    dfa = pd.DataFrame(f)
    df1 = df1.append(dfa, ignore_index=True)
```

```python
df2 = pd.DataFrame() # 2017, 2019, 2020, 2021년 -> 파일이 월별로 구성됨 & 망 변수가 추가됨 & 얘네끼리 형식 동일

for year in [2017, 2019, 2020, 2021]:
  for i in range(1,13):
    if year == 2021 and i==5:
      break
    f=pd.read_excel(f'/content/drive/MyDrive/Colab Notebooks/Bigdata_finedust/{year}년 {i}월.xlsx')
    dfa = pd.DataFrame(f)
    df2 = df2.append(dfa, ignore_index=True)
```

```python
df3 = pd.DataFrame() # 2018년 -> 망 변수가 추가된건 같은데 파일이 분기별로 구성됨, 2017, 2019, 2020, 2021년과 형식 동일(df2와 df3은 통합가능)

for i in range(1,5):
  f=pd.read_excel(f'/content/drive/MyDrive/Colab Notebooks/Bigdata_finedust/2018년 {i}분기.xlsx')
  dfa = pd.DataFrame(f)
  df3 = df3.append(dfa, ignore_index=True)
```

```python
df1721 = pd.concat([df2, df3] , ignore_index=True)
```

```python
df0.to_csv('/content/drive/MyDrive/Colab Notebooks/df0113.csv')
df1.to_csv('/content/drive/MyDrive/Colab Notebooks/df1416.csv')
df1721.to_csv('/content/drive/MyDrive/Colab Notebooks/df1721.csv')
# df3.to_csv('df3.csv', index=False)
```
