```python
#윤석열 명사추출
import pandas as pd
!pip install konlpy
from konlpy.tag import Okt
okt = Okt()
```

```python
df = pd.read_csv('/content/drive/MyDrive/빅데이터 기말과제 파일공유/트위터/윤석열/yoon_tweet.csv')
```

```python
df = df.fillna(' ')
df = df.drop_duplicates()
df
```

```python
import re
stop_words_path = '/content/drive/MyDrive/빅데이터 기말과제 파일공유/트위터/stop_words.txt'
with open(stop_words_path, encoding='utf-8') as f:
  stop_words = [w.strip() for w in f.readlines() if not w.isspace()]
def preprocessing(text):
  hangeul = re.compile(r'[^ㄱ-ㅣ가-힣]+')
  hangeuls = hangeul.sub('', text)
  nouns = okt.nouns(hangeuls)
  nouns = [noun for noun in nouns if len(noun)>1]
  nouns = [noun for noun in nouns if noun not in stop_words]
  return nouns
  
```

```python
df['nouns'] = df['content'].apply(lambda x: preprocessing(x))
df.to_csv('/content/drive/MyDrive/빅데이터 기말과제 파일공유/트위터/윤석열nouns.txt', index=False)
```
