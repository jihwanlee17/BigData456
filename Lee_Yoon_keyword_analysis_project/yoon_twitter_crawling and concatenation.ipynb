```python
import snscrape.modules.twitter as sntwitter
import pandas as pd
```

```python
# Creating list to append tweet data to
tweets_list2 = []

# Using TwitterSearchScraper to scrape data and append tweets to list
for i,tweet in enumerate(sntwitter.TwitterSearchScraper('윤석열 since:2021-11-01 until:2021-11-23').get_items()):
    if i>100000:
        break
    tweets_list2.append(tweet)
    
# Creating a dataframe from the tweets list above
tweets_df2 = pd.DataFrame(tweets_list2)
```

```python
tweets_df2.to_csv('/content/drive/MyDrive/윤석열/yoon_First Extract.csv')
```
